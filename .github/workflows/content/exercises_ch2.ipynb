## 演習1
2024年の次の値を使い質問に答えなさい。日本の実質GDPは`rgdp=557,434.1`（10億円），デフレーターは`deflator=109.3`（2015年=100），ドル円為替レートの平均は`ex = 151.37`円/ドルです。

### (1)
それぞれの変数を作成しなさい。

```python
rgdp = 557_434.1
deflator = 109.3
ex = 151.37
```

### (2)
名目GDPを計算し，変数`ngdp`に代入しなさい。

```python
ngdp = rgdp * deflator
ngdp
```

### (3)
名目GDPをドル換算し，変数`ngdp_dollar`に代入しなさい。

```python
ngdp_dollar = ngdp / ex
ngdp_dollar
```

## 演習2
2023年と2024年の消費者物価指数はそれぞれ`cpi23=105.6`と`cpi24=108.5`でです（2020年=100）。また，2023年と2024年のデフレーターはそれぞれ`def23=106.264`と`def24=109.319`です（2015年=100）

### (1)
それぞれの変数を作成しなさい。

```python
cpi23 = 105.6
cpi24 = 108.5
def23 = 106.264
def24 = 109.319
```

### (2)
消費者物価指数で測るインフレ率を％表示で計算しなさい。

```python
100 * (cpi24 - cpi23) / cpi23
```

### (3)
デフレーターで測るインフレ率を％表示で計算しなさい。

```python
100 * (def24 - def23) / def23
```

## 演習3
2024年の15歳以上人口`pop=10995`，労働力人口`labour=6957`，就業者`employed=6781`，完全失業者数`unemployed=176`です（単位：万人）。

### (3)
それぞれの変数を作成しなさい。

```python
pop = 10995
labour = 6957
employed = 6781
unemployed = 176
```

### (1)
労働力人口比率を％表示にし，変数`labour_rate`に代入しなさい。

```python
labour_rate = 100 * labour / pop
labour_rate
```

### (2)
就業率を％表示にし，変数`employed_rate`に代入しなさい。

```python
employed_rate = 100 * employed / pop
employed_rate
```

### (3)
完全失業率を％表示にし，変数`unemployed_rate`に代入しなさい。

```python
employed_rate = 100 * unemployed / labour
employed_rate
```

## 演習4
2024年の日本の失業率は`jp_unemployment = 2.57`%，アメリカの失業率は`us_unemployment = 4.03`%です。次の問いに答えなさい。

### (1)
変数を作成しなさい。

```python
jp_unemployment = 2.57
us_unemployment = 4.03
```

### (2)
日本の失業率がアメリカより低いかどうかを比較し，その結果を`is_japan_lower`（"Is Japanese unemployment rate lower?"の略）という変数に代入しなさい。    

```python
is_japan_lower = ( jp_unemployment < us_unemployment )
is_japan_lower
```

### (3)
日本の失業率が`3`%以下かどうかを比較し，その結果を`is_japan_below_3`（"Is Japanese unemployment rate below 3?"の略）という変数に代入しなさい。    

```python
is_japan_below_3 = ( jp_unemployment < 3 )
is_japan_below_3
```

### (4)
日本とアメリカの失業率が等しいかどうかを評価し，その結果を`is_equal`という変数に代入しなさい。

```python
is_equal = ( jp_unemployment == us_unemployment )
is_equal
```

## 演習5
ある国のインフレ率は`inflation = 1.5`%です。次の問いに答えなさい。

### (1)
変数を作成しなさい。

```python
inflation = 1.5
```

### (2)
インフレ率が`5`%上昇した場合，累積代入演算子を使用して`inflation`の値を更新しなさい。

```python
inflation += 5
inflation
```

### (3)
インフレ率が`3.5`%減少した場合、累積代入演算子を使用して`inflation`の値を更新しなさい。    

```python
inflation -= 3.5
inflation
```

### (4)
インフレ率が`2`倍になった場合、累積代入演算子を使用して`inflation`の値を更新しなさい。

```python
inflation *= 2
inflation
```
